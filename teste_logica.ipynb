```python
"""
when calls the function returns
a list with all relevant links 
in the refered page.
"""

def get_page_links(html):
    _temp_links= []
    _html_links= (
        BeautifulSoup(str(html),'html.parser')
        .find_all('a')
    )
    for i in _html_links[::2]:
        _formated_link= (
            str(i)
            .split('"')[1]
            .replace('../','')
        )
        _temp_links.append(_formated_link)

    return _temp_links
```
